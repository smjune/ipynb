### Q3

```python
df3= df.copy()
df3.columns
```

```python
df3.info()
```

```python
df3_1 = pd.get_dummies(df3, 'screen_size')
df3_1.shape
```

```python
col_list = df3_1.columns.tolist()
X_list = col_list.copy()
X_list.remove('sales')
X_list
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test  = train_test_split(df3_1, test_size=0.2, random_state=123)
train
```

```python
from sklearn.preprocessing import MinMaxScaler
myMMS = MinMaxScaler()
train.loc[:,:] = myMMS.fit_transform(train.loc[:,:])
test.loc[:,:] = myMMS.transform(test.loc[:,:])
train
```

```python
X_train = train.loc[:,X_list]
X_test = test.loc[:,X_list]
y_train = train['sales']
y_test = test['sales']
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
mse_list = []
for k in [3, 5, 7,9, 11]:
    myKNR = KNeighborsRegressor(n_neighbors=k)
    myKNR.fit(X_train, y_train)
    y_pred = myKNR.predict(X_test)
    mse_list.append(mean_squared_error(y_test, y_pred,squared=False))

pd.DataFrame(mse_list, index=[3, 5, 7,9, 11]).idxmin()
```
