>list - tuple - dict - set : 구분할 것    
list()  / tuple() / set()  / dict() 함수로 변환  

```python
total_wine = wine.columns.tolist()
numeric_wine = wine.describe().columns.tolist()
object_wine = list(set(total_wine) - set(numeric_wine))
object_wine
 
 # 모든 col set 에서 숫자형 col set 을 - 연산 = list(범주형 col set)
 # set 은 중복 제거됨
```
