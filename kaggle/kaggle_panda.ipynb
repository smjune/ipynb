````python
print(type(wine['taster_name'].str))
filt = wine['taster_name'].str.contains('Kerin', na=False)    # contains 는 NaN 을 처리할 수 없다.
filt2 = wine['variety'].str.lower().isin(['red blend'])       # isin([]) 으로 사용)

# wine[wine['taster_name'].str.contains('Kerin', na=Fales)] 는 왜 정상 wine, wine
# wine[filt & filt2].loc[:,'price'] >= 100 는 정상 bool dtype series
# wine[wine[filt & filt2].loc[:,'price'] >= 100] 는 오류 발생   (wine 과 wine[filt&filt2] 서로 다른 dataframe 이다.)
# wine[filt & filt2][wine[filt & filt2].loc[:,'price'] >= 100] 으로 진행 

# 다시 필터로 정리된 dataframe 을 다시 정의해서 filtered df 을 기준으로 다음 문제를 해결한다. (대상을 점점 좁혀간다. / 필요없는 col을 정리한다.) 
filtered_wine = wine[filt & filt2]
# filtered_wine[filtered_wine.loc[:,'price'] >= 100]    
# filtered_wine.loc[filtered_wine['price']>=100]
filtered_wine[filtered_wine.price >= 100]

# dataframe_filter[datafram_filter.loc['A']>100]   --> filtering 된 dataframe 이 같아야 한다. 
# df[0:-1][df[0:-1].loc["A"]>100] 혹은 df[df.A >100] 이렇게

# wine.drop(index = (filtered_wine.price >= 100).index)   # filtered된 인덱스를 전체 df에서 삭제
````
