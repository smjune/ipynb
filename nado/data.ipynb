OneHotEncoder 

my_OHE.fit 으로 train을 변환한 다음 동일한 train 기준으로 my_OHE.trainsform 을 로 test 도 변환할 수 있다. 

````python
from sklearn.preprocessing import OneHotEncoder

my_OHE = OneHotEncoder(sparse = False, drop='first')
df2_sh = my_OHE.fit_transform(df2[['학교']])

OHE= pd.DataFrame(df2_sh,
             columns=my_OHE.get_feature_names_out())
OHE
````

````python
from sklearn.compose import make_column_transformer


myt = make_column_transformer(
    (OneHotEncoder(sparse = False, drop='first'),['학교']),
    remainder='passthrough')

result = myt.fit_transform(df2)

pd.DataFrame(result,
             columns=['학교','이름',  '키', '국어', '영어', '수학', '과학', '사회', 'SW특기'])

````

````python
make_column_transformer((OneHotEncoder(sparse = False, drop='first'),['학교']))
````

ColumnTransformer(transformers = [])

````python
from sklearn.compose import ColumnTransformer

my_CT = ColumnTransformer(
    transformers=[('onehotencoder', OneHotEncoder(sparse = False, drop='first'), ['학교'])],
    remainder='passthrough')

df2_sh2 = my_CT.fit_transform(df2)

pd.DataFrame(df2_sh2,
             columns=['학교','이름',  '키', '국어', '영어', '수학', '과학', '사회', 'SW특기'])
````
