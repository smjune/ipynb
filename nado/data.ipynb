````python
from sklearn.compose import make_column_transformer


myt = make_column_transformer(
    (OneHotEncoder(sparse = False, drop='first'),['학교']),
    remainder='passthrough')

result = myt.fit_transform(df2)

pd.DataFrame(result,
             columns=['학교','이름',  '키', '국어', '영어', '수학', '과학', '사회', 'SW특기'])

````

ColumnTransformer(transformers = [])

````python
from sklearn.compose import ColumnTransformer

my_CT = ColumnTransformer(
    transformers=[('onehotencoder', OneHotEncoder(sparse = False, drop='first'), ['학교'])],
    remainder='passthrough')

df2_sh2 = my_CT.fit_transform(df2)

pd.DataFrame(df2_sh2,
             columns=['학교','이름',  '키', '국어', '영어', '수학', '과학', '사회', 'SW특기'])
````
