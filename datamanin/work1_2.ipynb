- Channel 데이터중 2021-10-03일 이후 각 채널의 처음 기록 됐던 구독자 수(subcnt)를 출력하라

```python
channel['ct'] = pd.to_datetime(channel['ct'])

over211003 = channel['ct']>='2021-10-03'

min_id = channel[over211003].groupby('channelid')['ct'].idxmin().values
min_id
# 먼저 2021-10-01 보다 큰 날짜 중 channelid 별로 ct 값의 최소값(처음 기록) 을 갖는 index list 을 축출
```

```python
# 해당 index (row) 의  channelname, subcnt 을 찾는다. 
channel.loc[min_id,['channelname','subcnt']]
```

```python
# groupby.apply()

def funx_subcnt(df_g):
    id = df_g['ct'].idxmin()
    return df_g.loc[id,['channelname','subcnt']]

channel[over211003].groupby('channelid')[['ct','channelname','subcnt']].apply(funx_subcnt)
```

- 각채널의 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00 까지 구독자수 (subcnt) 의 증가량을 구하여라

```python
date = (channel['ct']>='2021-10-03 03:00:00')&(channel['ct']<='2021-11-01 15:00:00')
```

```python
# apply function 확인 
# channel 대신 groupby key 을 적용하면 된다.

start = channel.loc[date,'ct'].idxmin()
stop = channel.loc[date,'ct'].idxmax()
channel.loc[start,'subcnt'] - channel.loc[stop,'subcnt']
```

```python
def FunX(g_df):
    # print(g_df.name)
    start = g_df.loc[date,'ct'].idxmin()
    stop = g_df.loc[date,'ct'].idxmax()
    return g_df.loc[stop,'subcnt'] - g_df.loc[start,'subcnt']
# groupby key 별로 결과가 리턴된다.

channel.groupby('channelname')[['ct','subcnt']].apply(FunX)
```
